```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

In this workspace, you will work with the fuel economy dataset from the previous lesson on bivariate plots.

```python
# read in dataframe
fuel_econ = pd.read_csv('fuel-econ.csv')
fuel_econ.head()
```

**Task 1**: Plot the city ('city') vs. highway ('highway') fuel efficiencies (both in mpg) for each vehicle class ('VClass'). Don't forget that vehicle class is an ordinal variable with levels {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, Large Cars}.

```python
# convert VClass column to categorical datatype

sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses);
```

```python
g = sns.FacetGrid(data = fuel_econ, col = 'VClass', height = 5, col_wrap = 3)
g.map(plt.scatter, 'city', 'highway', alpha = 0.1);
```

**Task 2**: Plot the relationship between engine size ('displ', in meters), vehicle class, and fuel type ('fuelType'). For the lattermost feature, focus only on Premium Gasoline and Regular Gasoline cars. What kind of relationships can you spot in this plot?

```python
# create new column dropping the word 'gasoline' from fuel type
fuel_econ['fuel'] = fuel_econ['fuelType'].apply(lambda x: x.split()[0])
# just using regular or premium gasoline for this sub-dataframe
fuel_econ_sub = fuel_econ[fuel_econ['fuel'].isin(['Regular', 'Premium'])]
```

```python
# boxplot of vehicle class and engine size clustered by fuel type
# use hue_order to show regular fuel first
# use bbox_to_anchor to place legend outside plot
ax = sns.boxplot(data = fuel_econ_sub, x = 'VClass', y = 'displ', hue = 'fuel', hue_order = ['Regular', 'Premium']);
ax.legend(loc = 6, bbox_to_anchor = (1.0, 0.5), title = 'Fuel Type');
plt.xticks(rotation = 15);
plt.xlabel('Vehicle Class');
plt.ylabel('Displacement (l)');
```

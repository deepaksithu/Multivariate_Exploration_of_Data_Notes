```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

In this notebook, you'll be working with the Pokémon dataset from the univariate plots lesson.

```python
# read in dataframe
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

**Task 1**: To start, let's look at the relationship between the Pokémon combat statistics of Speed, Defense, and Special-Defense. If a Pokémon has higher defensive statistics, does it necessarily sacrifice speed? Create a single plot to depict this relationship.

```python
# get default figure size (print fig_size)
fig_size = plt.rcParams['figure.figsize']
#print(fig_size) (6, 4)

# set figure width and height
fig_size[0] = 12
fig_size[1] = 8
# set new figure size for plot
plt.rcParams['figure.figsize'] = fig_size

# scatter plot of defense and special defense with color denoting speed
# also with axis limits to improve visibility 
plt.scatter(data = pokemon, x = 'defense', y = 'special-defense', c = 'speed',
           cmap = 'viridis');
plt.colorbar(label = 'Speed');
plt.xlabel('Defense')
plt.ylabel('Special Defense')
plt.xlim(0, 200)
plt.ylim(0, 175);
```

To complete the second task, we need to first reshape the dataset so that all Pokémon types are recorded in a single column. This will add duplicates of Pokémon with two types, which is fine for the task to be performed.

```python
type_cols = ['type_1','type_2']
non_type_cols = pokemon.columns.difference(type_cols)
pkmn_types = pokemon.melt(id_vars = non_type_cols, value_vars = type_cols, 
                          var_name = 'type_level', value_name = 'type').dropna()
pkmn_types.head()
```

## Version 1 with sns.regplot

```python
# list of pokemon types for sub-dataframe
i = ['fairy', 'dragon']
# create sub-dataframe
fd_pkmn = pkmn_types[pkmn_types['type'].isin(i)]
```

```python
# list of lists for categorical, non-positional variable pokemon type
ttype_markers = [['fairy', 'o'],
                ['dragon', 'x']]

# looping through list elements
# use ttype to select only the plots with the matching pokemon type
# set the marker with marker in sns.regplot
for ttype, marker in ttype_markers:
    plot_data = fd_pkmn.loc[fd_pkmn['type'] == ttype]
    ax = sns.regplot(data = plot_data, x = 'weight', y = 'height',
           fit_reg = False, marker = marker)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend(['Fairy', 'Dragon'], title = 'Pokemon Types')
# set log scale for x-axis
ax.set_xscale('log');
# create list of x-ticks based on log scale
x_ticks = [0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000]
# set x-tick placement and label
plt.xticks(ticks = x_ticks, labels = x_ticks);
```

## Version 2 with sns.FacetGrid and plt.scatter

```python
# create sub-dataframe of just fairy and dragon pokemon
pokemon_sub = pkmn_types.loc[pkmn_types['type'].isin(['fairy', 'dragon'])]
```

```python
#  set up FacetGrid to show pokemon type in as color
g = sns.FacetGrid(data = pokemon_sub, hue = 'type', height = 5, aspect = 1.5)
# scatterplot of weight and height
g.map(plt.scatter, 'weight', 'height')
# set scale of x axis to log
g.set(xscale = 'log')
# create log scale list of tick values for x-axis
x_ticks = [0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000]
# set the values and labels of the ticks to the list
g.set(xticks = x_ticks, xticklabels = x_ticks)
# add legend (messing with this from default does weird things!)
g.add_legend();
```
